```python
import pandas as pd

temporal_columns = [
    'transaction_date',
    'last_purchase_date',
    'product_expiry_date',
    'product_manufacture_date',
    'promotion_start_date',
    'promotion_end_date'
]

# Ensure the temporal columns exist in the DataFrame before processing
existing_temporal_columns = [col for col in temporal_columns if col in data.columns]

# Convert temporal columns to datetime objects
# Use errors='coerce' to turn any unparseable dates into NaT (Not a Time)
for col in existing_temporal_columns:
    data[col] = pd.to_datetime(data[col], errors='coerce')

print("Minimum and Maximum Dates for Temporal Columns:")
for col in existing_temporal_columns:
    min_date = data[col].min()
    max_date = data[col].max()
    print(f"\n--- {col} ---")
    print(f"Min Date: {min_date}")
    print(f"Max Date: {max_date}")
```

```python
# Define temporal attributes to remove
temporal_columns = [
    'transaction_date',
    'last_purchase_date',
    'product_expiry_date',
    'product_manufacture_date',
    'promotion_start_date',
    'promotion_end_date'
]

# Remove temporal attributes
data = data.drop(columns=[col for col in temporal_columns if col in data.columns])

# Verify columns after removal
print("Columns after removing temporal attributes:", data.columns.tolist())
```

## Descriptive Statistics on all attributes

```python
# Display all columns when printing DataFrame
pd.set_option('display.max_columns', None)


import pandas as pd

# Assuming your DataFrame is named 'data'
# Get all columns from the DataFrame
all_columns = data.columns.tolist()

# Verify the number of columns
print(f"Total columns in DataFrame: {len(all_columns)}")
print("Columns:", all_columns)

from sklearn.model_selection import train_test_split
X_train, X_test, _, _ = train_test_split(data, data.index, test_size=0.3, random_state=42)

# Generate descriptive statistics for numerical columns (int64, float64)
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
numerical_stats = data[numerical_columns].describe()

# Generate descriptive statistics for categorical columns (object)
categorical_columns = data.select_dtypes(include=['object']).columns
categorical_stats = data[categorical_columns].describe()

# Print the statistics
print("\nNumerical Columns Statistics:")
print(numerical_stats)
print("\nCategorical Columns Statistics:")
print(categorical_stats)
```

## Sampling to 20,000 Records - Simple Random Sampling

```python
import pandas as pd

# Check if data has more than or equal to 20000 rows
sample_size = min(20000, len(data))  # Sample 20,000 rows

# Sample the data
sampled_data = data.sample(n=sample_size, random_state=42)

# Display the sampled data
print(sampled_data)
```
