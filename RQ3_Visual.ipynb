```python
import matplotlib.pyplot as plt
import numpy as np

# First visualization: Recall comparison across all models
models = [
    'DT - No Bal - All Feat', 'DT - No Bal - Sel Feat',
    'DT - Bal - All Feat', 'DT - Bal - Sel Feat',
    'KNN - No Bal - All Feat', 'KNN - No Bal - Sel Feat',
    'KNN - Bal - All Feat', 'KNN - Bal - Sel Feat',
    'RF - No Bal - All Feat', 'RF - No Bal - Sel Feat',
    'RF - Bal - All Feat', 'RF - Bal - Sel Feat'
]
recalls = [0.53, 0.53, 0.55, 0.55, 0.67, 0.64, 0.68, 0.63, 0.70, 0.66, 0.72, 0.71]

plt.figure(figsize=(12, 6))
bars = plt.bar(models, recalls, color=['#FF9999' if 'DT' in m else '#66B2FF' if 'KNN' in m else '#99FF99' for m in models])
plt.xlabel('Model Configurations')
plt.ylabel('Recall')
plt.title('Cluster 0 Recall Comparison Across 12 Models')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 0.8)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
import numpy as np

# Data for Cluster 0
metrics = ['Precision', 'Recall', 'F1-score', 'Specificity']
original = [0.74, 0.72, 0.73, 0.8713]  # Cluster 0 metrics from original model
improved = [0.76, 0.84, 0.80, 0.8353]  # Cluster 0 metrics from improved model

# Set up the plot
x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(8, 6))
plt.bar(x - width/2, original, width, label='Original RF - Bal - All Feat', color='#F60029')
plt.bar(x + width/2, improved, width, label='Improved RF - Bal - All Feat', color='#782170')

# Customize the plot
plt.xlabel('Cluster 0 Metrics')
plt.ylabel('Score')
plt.title('Cluster 0 Performance: Original vs Improved Random Forest (Balanced, All Features)')
plt.xticks(x, metrics, rotation=45, ha='right')
plt.ylim(0, 1.0)
plt.legend()

# Add value labels on top of bars
for i, v in enumerate(original):
    plt.text(i - width/2, v + 0.01, f'{v:.4f}', ha='center', va='bottom')
for i, v in enumerate(improved):
    plt.text(i + width/2, v + 0.01, f'{v:.4f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
